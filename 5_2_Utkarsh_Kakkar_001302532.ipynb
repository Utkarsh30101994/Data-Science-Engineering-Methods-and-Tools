```python
import scipy.stats
import numpy as np
data = scipy.stats.norm.rvs(size=100000, loc=0, scale=1.5, random_state=123)
hist = np.histogram(data, bins=200)
hist_dist = scipy.stats.rv_histogram(hist)
```

```python
len(data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("data with normal distribution")
plt.plot(data[0:1000], label='data')
plt.show()
```

```python
hist_dist.pdf(1.0)
hist_dist.cdf(2.0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
X = np.arange(-5.0, 5.0, 4500)
plt.title("histogram and PDF from Template")
plt.hist(data, density=True, bins=1000)
plt.plot(X, hist_dist.pdf(X), label='PDF')
plt.show()
```
